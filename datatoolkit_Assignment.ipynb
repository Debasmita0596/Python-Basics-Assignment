```python
#6.Create a 3D NumPy array of shape (3, 3, 3) containing random integers between 1 and 10. Perform the following operations:

 #a) Find the indices of the maximum values along each depth level (third axis).

 #b) Perform element-wise multiplication of between both array

 #Here is how to create a 3D Numpy array of shape (3,3,3) containing random integers between 1& 10,find the indices of the maximum values along the third axis,and then perform element-wise multiplication of the original array with the maximum values:



 #import numpy as np

# Create a 3D array with random integers between 1 and 10
array_3d = np.random.randint(1, 11, size=(3, 3, 3))

# Find the indices of the maximum values along the third axis
max_indices = np.argmax(array_3d, axis=2)

# Create an array of the maximum values along the third axis
max_values = np.max(array_3d, axis=2)

# Perform element-wise multiplication
result = array_3d * max_values[:, :, np.newaxis]  # Broadcasting max_values to match the shape of array_3d

# Output the results
print("3D Array:")
print(array_3d)
print("\nIndices of maximum values along each depth level:")
print(max_indices)
print("\nMaximum values along each depth level:")
print(max_values)
print("\nElement-wise multiplication result:")
print(result)

#Explanation:
#Creating the Array: We create a random 3D array with integers between 1 and 10.
#Finding Indices of Maximum Values: We use np.argmax() to get the indices of the maximum values along the third axis.
#Extracting Maximum Values: We use np.max() to get the actual maximum values.
#Element-wise Multiplication: We multiply the original array by the maximum values, utilizing broadcasting to align the dimensions.
```

```python
#7.Clean and transform the 'Phone' column in the sample dataset to remove non-numeric characters and convert it to a numeric data type. Also display the table attributes and data types of each column
#To clean and transform a 'Phone' column in a dataset to remove non-numeric characters and convert it to a numeric data type, you can use pandas in Python. Below is a sample implementation, including how to display the attributes and data types of each column:


import pandas as pd

# Sample dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Phone': ['123-456-7890', '(234) 567-8901', '345.678.9012'],
    'Age': [25, 30, 35]
}

df = pd.DataFrame(data)

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Clean and transform the 'Phone' column
df['Phone'] = df['Phone'].replace(r'\D+', '', regex=True).astype(int)

# Display the cleaned DataFrame
print("\nCleaned DataFrame:")
print(df)

# Display attributes and data types of each column
print("\nData Types of Each Column:")
print(df.dtypes)

# Display the DataFrame's attributes
print("\nDataFrame Attributes:")
print(df.info())

#Explanation:
#Data Creation: A sample dataset is created using a dictionary and converted to a DataFrame.
#Cleaning the 'Phone' Column: The replace() function with a regex pattern \D+ removes all non-numeric characters, and astype(int) converts the cleaned strings to integers.
#Displaying Data Types: df.dtypes shows the data type of each column.
#DataFrame Attributes: The info() method displays the DataFrame's summary, including the number of entries, column names, and data types
```

```python
#18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'
#To create a line chart of a sine wave using Bokeh, you can use the following code:



from math import pi, sin
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare data
x = np.linspace(0, 2 * pi, 100)
y = np.sin(x)

# Create a new plot with a title and axis labels
p = figure(title="Sine Wave Function", x_axis_label='X', y_axis_label='sin(X)', width=800, height=400)

# Add grid lines
p.grid.grid_line_alpha = 0.3

# Add a line renderer
p.line(x, y, legend_label="sin(x)", line_width=2)

# Show the plot
output_notebook()  # Use this in Jupyter notebooks
show(p)
#Explanation:
#Data Preparation: The sine wave is created using numpy to generate values from 0 to 2π.
#Figure Creation: A new figure is created with figure() where you can set the title and labels for the axes.
#Grid Lines: The grid lines are customized for better visibility.
#Line Renderer: The sine wave is plotted using the line() method.
#Displaying the Plot: Finally, show() is called to display the plot.
```

```python
#19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
#Bar Chart

#Here's how you can generate a bar chart with randomly generated categorical data using Bokeh, including colored bars based on their values and hover tooltips:


import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool

# Generate random categorical data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(1, 100, size=len(categories))

# Create a DataFrame for better handling
data = pd.DataFrame({'categories': categories, 'values': values})

# Define a color mapping based on the values
colors = ['#%02x%02x%02x' % (int(value * 2.55), 0, 0) for value in values]  # Red shades based on value

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(categories=data['categories'], values=data['values'], colors=colors))

# Create a new plot with title and axis labels
p = figure(x_range=data['categories'], title="Random Categorical Bar Chart",
           x_axis_label='Categories', y_axis_label='Values', height=400)

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Add bars
p.vbar(x='categories', top='values', width=0.9, source=source, color='colors')

# Show the plot
output_notebook()  # Use this in Jupyter notebooks
show(p)
#Explanation:
#Data Generation: Random values are generated for five categories.
#Color Mapping: Colors are created based on the values, with darker shades for higher values.
#ColumnDataSource: This helps in managing the data and tooltips.
#Figure Creation: A figure is created with specified title and axis labels.
#Hover Tool: The hover tool is set up to display category names and values when hovering over bars.
#Vertical Bars: The bars are drawn using vbar(), and colors are applied from the data source.
#Display: Finally, the plot is displayed using show().
```

```python
#20.Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as 'Simple Line Plot'

#Here's how you can create a basic line plot using Plotly with a randomly generated dataset:



import numpy as np
import plotly.graph_objects as go

# Generate random data
x = np.linspace(0, 10, 100)
y = np.random.rand(100)

# Create the line plot
fig = go.Figure()

# Add a line trace
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data'))

# Update layout with title and axis labels
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X Axis',
    yaxis_title='Y Axis'
)

# Show the plot
fig.show()
#Explanation:
#Data Generation: Random values are generated for the y-axis, while x-values are evenly spaced between 0 and 10.
#Figure Creation: A Figure object is created using Plotly.
#Line Trace: A line trace is added using go.Scatter() with mode='lines'.
#Layout Update: The layout is updated to include the title and axis labels.
#Display: Finally, the plot is displayed using fig.show().
```

```python
#17.Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

#To create a Seaborn scatter plot with the specified features, we can use the following Python code. This code generates two random arrays, colors the points based on their quadrants, and includes a legend, axis labels, and a title.


import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(0)  # For reproducibility
x = np.random.randn(100)
y = np.random.randn(100)

# Create a DataFrame to hold the data and assign quadrants
data = pd.DataFrame({'x': x, 'y': y})

def assign_quadrant(row):
    if row['x'] >= 0 and row['y'] >= 0:
        return 'Quadrant I'
    elif row['x'] < 0 and row['y'] >= 0:
        return 'Quadrant II'
    elif row['x'] < 0 and row['y'] < 0:
        return 'Quadrant III'
    else:
        return 'Quadrant IV'

data['Quadrant'] = data.apply(assign_quadrant, axis=1)

# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='x', y='y', hue='Quadrant', palette='deep')
plt.title('Quadrant-wise Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(title='Quadrants')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.xlim(-3, 3)
plt.ylim(-3, 3)
plt.show()


#Explanation:
#Data Generation: Two arrays of random numbers are created using numpy.
#Quadrant Assignment: A function determines which quadrant each point belongs to based on its x and y values.
#Plotting: The scatter plot is created using Seaborn, with points colored according to their quadrant. The axes are labeled, a title is added, and a grid is included for clarity. Vertical and horizontal lines represent the axes.
```
