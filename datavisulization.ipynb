```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
#To generate a synthetic dataset with correlated features and visualize the correlation matrix using a heatmap, you can use NumPy to create the dataset and Seaborn along with Matplotlib for visualization. Here's how to do this in Python:

#Step-by-Step Code
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate synthetic dataset with correlated features
np.random.seed(0)  # For reproducibility
num_samples = 100

# Generate random data
x = np.random.rand(num_samples)  # Feature 1
y = 2 * x + np.random.normal(0, 0.1, num_samples)  # Feature 2, correlated with Feature 1
z = -x + np.random.normal(0, 0.1, num_samples)  # Feature 3, inversely correlated with Feature 1

# Create a DataFrame
data = pd.DataFrame({
    'Feature 1': x,
    'Feature 2': y,
    'Feature 3': z
})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title('Correlation Matrix Heatmap')
plt.show()
#Explanation:
#Synthetic Data Generation:

#A random dataset with three features (Feature 1, Feature 2, and Feature 3) is generated.
#Feature 1 is generated using np.random.rand(num_samples).
#Feature 2 is created to have a positive correlation with Feature 1 by scaling x and adding some noise.
#Feature 3 is created to have an inverse correlation with Feature 1 by negating x and adding noise.
#DataFrame Creation:

#The features are stored in a Pandas DataFrame for easy manipulation and correlation calculation.
#Correlation Matrix Calculation:

#The correlation matrix is calculated using the data.corr() method.
#Heatmap Visualization:

#A heatmap is created using sns.heatmap(), with the annot=True parameter to display correlation values on the heatmap.
#The cmap='coolwarm' parameter sets the color palette for the heatmap.
#The square=True parameter ensures that the heatmap cells are square-shaped.
#Display: Finally, plt.show() renders the heatmap.
```

```python
#4. Using the given x and y data, generate a 3D surface plot to visualize the function z =sin (√x2=y2)
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the range and meshgrid for x and y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Define the function
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create the plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Add labels and a color bar
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)

plt.show()

#This code will produce a smooth 3D surface plot for the function, showing the wave pattern generated by the sine function. The color map highlights variations in the z-values, making the peaks and valleys more distinct
```

```python
#3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

#fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
#counts = [20, 25, 30, 35]

#To create a Bokeh bar chart for the given dataset, I'll use the fruits and counts lists to generate the chart. Here’s the code to produce a simple bar chart with Bokeh:
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource
from bokeh.palettes import Spectral6

# Define the data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a figure
p = figure(x_range=fruits, height=400, title="Fruit Counts",
           toolbar_location=None, tools="")

# Add bars to the chart
p.vbar(x='fruits', top='counts', width=0.6, source=source,
       line_color='white', fill_color=factor_cmap('fruits', palette=Spectral6, factors=fruits))

# Customize the chart
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.yaxis.axis_label = "Count"
p.xaxis.axis_label = "Fruit"

# Display the chart
output_notebook()
show(p)
#his script will generate a colorful bar chart with distinct colors for each fruit, displaying the counts vertically. Each fruit category is labeled on the x-axis, and the count values are on the y-axis
```

```python
#4. Create a Bokeh histogram to visualize the distribution of the given data.

#data_hist = np.random.randn(1000)
#hist, edges = np.histogram(data_hist, bins=30)

#To create a histogram with Bokeh using the provided data, I'll use data_hist and its histogram bins hist and edges. Here’s the code to create and display the histogram in Bokeh:

import numpy as np
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource

# Generate data and histogram values
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Set up the data source for Bokeh
source = ColumnDataSource(data=dict(
    top=hist,
    left=edges[:-1],
    right=edges[1:]
))

# Create the figure
p = figure(title="Histogram of Data Distribution", height=400, width=600)
p.quad(top='top', bottom=0, left='left', right='right', source=source, fill_color="navy", line_color="white", alpha=0.7)

# Customize the chart
p.yaxis.axis_label = "Frequency"
p.xaxis.axis_label = "Value"
p.grid.grid_line_color="gray"

# Display the chart
output_notebook()
show(p)

#This code will produce an interactive Bokeh histogram showing the distribution of the data in data_hist. The histogram will have 30 bins, and each bin’s height corresponds to the frequency of values within that range.
```

```python
#5.Create a Bokeh heatmap using the provided dataset.
#data_heatmap = np.random.rand(10, 10)
#x = np.linspace(0, 1, 10)
#y = np.linspace(0, 1, 10)
#xx, yy = np.meshgrid(x, y)

#To create a heatmap in Bokeh with the given dataset, we’ll treat each cell in data_heatmap as an individual rectangle and assign colors based on its value. Here’s how to achieve this using the 10x10 grid data in Bokeh:

import numpy as np
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColorBar, LinearColorMapper, ColumnDataSource
from bokeh.transform import linear_cmap

# Generate the data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Prepare data for Bokeh's ColumnDataSource
source = ColumnDataSource(data=dict(
    x=xx.flatten(),
    y=yy.flatten(),
    value=data_heatmap.flatten()
))

# Set up the color mapper
color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())

# Create the figure
p = figure(title="Heatmap", x_axis_label="X", y_axis_label="Y", tools="", toolbar_location=None,
           x_range=(0, 1), y_range=(0, 1), width=500, height=500)

# Create the heatmap using the `rect` glyph
p.rect(x="x", y="y", width=0.1, height=0.1, source=source,
       fill_color=linear_cmap('value', 'Viridis256', data_heatmap.min(), data_heatmap.max()), line_color=None)

# Add a color bar
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Display the plot
output_notebook()
show(p)


```
