1. What are the five key concepts of Object-Oriented Programming (OOP)?

The five key concepts of Object-Oriented Programming (OOP) are:

Class: A blueprint for creating objects. It defines a datatype by bundling data (attributes) and methods (functions) that work on the data into a single unit.

Object: An instance of a class. It represents a specific implementation of a class with actual data, and through it, you can interact with the methods and attributes defined in the class.

Encapsulation: The concept of restricting access to certain components of an object. This is achieved by bundling the data and the methods that manipulate the data together and using access modifiers like private, protected, and public to control access.

Inheritance: The mechanism by which one class (child or subclass) inherits properties and behaviors (methods) from another class (parent or superclass). This allows code reusability and the creation of a hierarchical relationship between classes.

Polymorphism: The ability to present the same interface or behavior in different forms. This can be achieved through method overloading (same method name with different parameters) or method overriding (subclass changes the implementation of a method defined in a superclass).

These concepts form the foundation of OOP and enable the creation of modular, reusable, and maintainable code.

```python
#3. Explain the difference between instance methods and class methods. Provide an example of each.

#In Python, instance methods and class methods serve different purposes and have distinct behaviors. Here's an explanation of the differences, followed by examples.

#Instance Methods

#Definition: Instance methods are the most common type of methods in a class. They operate on an instance of the class, meaning they require an object to be called.

#Access: They can access and modify instance-specific data (attributes) and can call other instance methods.

#Parameter: The first parameter of an instance method is always self, which refers to the specific instance of the class that is calling the method.

#Class Methods

#Definition: Class methods are methods that operate on the class itself rather than on an instance of the class. They are marked with the @classmethod decorator.

#Access: They can only access class variables (variables shared across all instances), but they cannot modify instance-specific data.

#Parameter: The first parameter of a class method is cls, which refers to the class itself, not a particular instance.

#Example

class Car:
    # Class variable
    total_cars = 0

    def __init__(self, make, model, year):
        # Instance variables
        self.make = make
        self.model = model
        self.year = year
        Car.total_cars += 1  # Increment total cars whenever a new car is created

    # Instance method
    def display_info(self):
        print(f"Car Information: {self.year} {self.make} {self.model}")

    # Class method
    @classmethod
    def display_total_cars(cls):
        print(f"Total cars created: {cls.total_cars}")

# Example usage
car1 = Car("Toyota", "Corolla", 2020)
car1.display_info()  # Calling instance method

car2 = Car("Honda", "Civic", 2021)

# Calling class method without an instance
Car.display_total_cars()

#Explanation:

#Instance Method (display_info):
#Operates on an individual Car object (car1 in the example).
#Accesses the instance-specific attributes (make, model, and year).
#Class Method (display_total_cars):
#Operates on the class itself (Car).
#Accesses the class variable total_cars, which is shared across all instances.
#Can be called either from an instance or directly from the class (Car.display_total_cars()).
#Output:

#Car Information: 2020 Toyota Corolla
#Total cars created: 2

```

```python
#5. What are the three types of access modifiers in Python? How are they denoted?

#In Python, access modifiers control the accessibility or visibility of class members (attributes and methods). Although Python doesn't have strict access control like some other languages (e.g., private, protected, and public in Java or C++), it uses naming conventions to define three levels of access: public, protected, and private.

#1. Public Access Modifier
#Definition: Public members are accessible from anywhere, both inside and outside the class.
#Denotation: Public attributes and methods are denoted by no leading underscores in their names.

class Car:
    def __init__(self, make, model):
        self.make = make  # Public attribute
        self.model = model  # Public attribute

    def display_info(self):  # Public method
        print(f"Car: {self.make} {self.model}")

car = Car("Toyota", "Corolla")
print(car.make)  # Accessible from outside the class
car.display_info()  # Accessible from outside the class

#2. Protected Access Modifier
#Definition: Protected members are intended to be accessed only within the class and its subclasses. However, in Python, they can still be accessed from outside (though it's a convention to treat them as non-public).
#Denotation: Protected members are denoted by a single leading underscore (_).

class Car:
    def __init__(self, make, model):
        self._engine_type = "V6"  # Protected attribute

    def _display_engine_info(self):  # Protected method
        print(f"Engine Type: {self._engine_type}")

car = Car("Toyota", "Corolla")
print(car._engine_type)  # Can be accessed, but convention suggests not to
car._display_engine_info()  # Can be called, but treated as non-public
#3. Private Access Modifier
#Definition: Private members are intended to be accessible only within the class. Python achieves this through name mangling (the interpreter changes the name of the attribute or method to make it harder to access from outside).
#Denotation: Private members are denoted by a double leading underscore (__).

class Car:
    def __init__(self, make, model):
        self.__max_speed = 200  # Private attribute

    def __display_max_speed(self):  # Private method
        print(f"Max Speed: {self.__max_speed} km/h")

car = Car("Toyota", "Corolla")
# print(car.__max_speed)  # This will raise an AttributeError
# car.__display_max_speed()  # This will also raise an AttributeError

# Accessing private members using name mangling
print(car._Car__max_speed)  # This will work due to name mangling
car._Car__display_max_speed()  # This will also work due to name mangling
#Key Points:
#Public members are accessible anywhere.
#Protected members (single underscore _) are meant for internal use, but can still be accessed outside the class. It's a convention, not strict enforcement.
#Private members (double underscore __) are modified through name mangling, making them harder to access externally, but still accessible using the mangled name (_ClassName__attribute).
#Output for Private Example:

200
#Max Speed: 200 km/h

```

```python
#6. Describe the five types of inheritance in Python. Provide a simple example of multiple inheritance.

#In Python, inheritance allows a class to inherit properties and behaviors (attributes and methods) from another class. There are five types of inheritance, each describing different inheritance structures:

#1. Single Inheritance
#A single class inherits from one parent class.

class Animal:
    def speak(self):
        print("Animal sound")

class Dog(Animal):
    pass

dog = Dog()
dog.speak()  # Inherits from Animal

#2. Multiple Inheritance
#A class inherits from more than one parent class. This allows the child class to inherit methods and attributes from all its parent classes.

class Engine:
    def start(self):
        print("Engine started")

class Wheels:
    def roll(self):
        print("Wheels are rolling")

class Car(Engine, Wheels):
    pass

my_car = Car()
my_car.start()  # Inherits from Engine
my_car.roll()   # Inherits from Wheels

#3. Multilevel Inheritance
#A class inherits from a parent class, which itself is a child of another class (i.e., a class inherits from a class that already inherited from another class).

class Animal:
    def speak(self):
        print("Animal sound")

class Mammal(Animal):
    def walk(self):
        print("Mammal walking")

class Dog(Mammal):
    def bark(self):
        print("Dog barking")

dog = Dog()
dog.speak()  # Inherits from Animal
dog.walk()   # Inherits from Mammal
dog.bark()   # Defined in Dog

#4. Hierarchical Inheritance
#Multiple child classes inherit from the same parent class.

class Animal:
    def speak(self):
        print("Animal sound")

class Dog(Animal):
    def bark(self):
        print("Dog barking")

class Cat(Animal):
    def meow(self):
        print("Cat meowing")

dog = Dog()
cat = Cat()
dog.speak()  # Inherited from Animal
cat.speak()  # Inherited from Animal

#5. Hybrid Inheritance
#A combination of two or more types of inheritance (e.g., a mix of multilevel and multiple inheritance). Python supports hybrid inheritance, and it uses a method resolution order (MRO) to determine the order in which methods are inherited when multiple paths are involved.
#Example of Multiple Inheritance:

class Mother:
    def speak(self):
        print("Mother speaking")

class Father:
    def walk(self):
        print("Father walking")

class Child(Mother, Father):
    def play(self):
        print("Child playing")

# Creating an instance of Child
child = Child()
child.speak()  # Inherited from Mother
child.walk()   # Inherited from Father
child.play()   # Defined in Child

#Output:

#Mother speaking
#Father walking
#Child playing


```

```python
#7. What is the Method Resolution Order (MRO) in Python? How can you retrieve it programmatically?

#Method Resolution Order (MRO) in Python

#The Method Resolution Order (MRO) defines the order in which Python looks for a method or attribute in a hierarchy of classes when dealing with inheritance, especially in cases involving multiple inheritance. MRO ensures that the right method is called by resolving any potential ambiguity in the inheritance tree.

#Python uses the C3 Linearization algorithm (also known as the C3 superclass linearization) to determine the MRO. This ensures a consistent and predictable order of method resolution, avoiding potential conflicts in class hierarchies.

#MRO Rules:
#The class itself is checked first.
#Then the parent classes are checked in the order they are inherited.
#Python proceeds depth-first, left-to-right (in cases of multiple inheritance).
#If a class is repeated in the inheritance hierarchy, it is only checked once (following the C3 linearization rules).
#Retrieving the MRO Programmatically
#You can retrieve the MRO for any class using:

#The __mro__ attribute, which returns a tuple of the classes in the MRO.
#The built-in mro() method of a class, which returns a list of classes in the resolution order.
#Example:

class A:
    def method(self):
        print("Method in A")

class B(A):
    def method(self):
        print("Method in B")

class C(A):
    def method(self):
        print("Method in C")

class D(B, C):
    pass

# Retrieving MRO using __mro__ attribute
print(D.__mro__)

# Retrieving MRO using the mro() method
print(D.mro())

# Example of calling a method
d = D()
d.method()  # MRO determines which method to call
#Output:

#(<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>)
#[<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>]
#Method in B

#Explanation:
#MRO of Class D: According to the MRO, the search order is D -> B -> C -> A -> object. When d.method() is called, Python checks D first, then B (where it finds method()), so it calls B's method.
#__mro__ and mro(): Both of these methods return the same MRO sequence, showing the order in which Python looks for methods.
#Importance of MRO:
#Consistency: It ensures that classes are searched in a consistent order when multiple inheritance is involved.
#Conflict resolution: It avoids issues like the diamond problem, where a class could inherit from two classes that share a common ancestor. MRO ensures that each class is visited only once.


```

```python
#8. Create an abstract base class `Shape` with an abstract method `area()`. Then create two subclasses`Circle` and `Rectangle` that implement the `area()` method.

#In Python, abstract base classes can be created using the abc module. An abstract base class defines methods that must be implemented by its subclasses. The abstract methods act as placeholders and must be implemented in the derived classes.

#Here’s how you can define an abstract base class Shape with an abstract method area(), and implement it in two subclasses Circle and Rectangle.

#Example:

from abc import ABC, abstractmethod
import math

# Abstract base class
class Shape(ABC):

    @abstractmethod
    def area(self):
        pass  # Abstract method, must be implemented by subclasses

# Subclass Circle implementing the abstract method area()
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * (self.radius ** 2)

# Subclass Rectangle implementing the abstract method area()
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# Example usage
circle = Circle(5)
rectangle = Rectangle(4, 6)

print(f"Area of Circle: {circle.area()}")      # Output: Area of Circle: 78.53981633974483
print(f"Area of Rectangle: {rectangle.area()}")  # Output: Area of Rectangle: 24

#Explanation:

#Abstract Base Class Shape:

#The Shape class inherits from ABC (Abstract Base Class), which is provided by the abc module.
#The @abstractmethod decorator marks area() as an abstract method, meaning it must be implemented by any subclass of Shape.

#Circle and Rectangle Subclasses:

#Both Circle and Rectangle inherit from Shape and implement the area() method as per their respective formulas.

#In Circle, the area is calculated using the formula 𝜋𝑟2

#In Rectangle, the area is calculated as width × height

#Usage:

#The area() method is called on both the Circle and Rectangle objects to compute their areas.

#Output:
#Area of Circle: 78.53981633974483
#Area of Rectangle: 24
#This structure follows the principles of object-oriented programming, enforcing that all subclasses of Shape must define their own area() method.
```

```python
#9. Demonstrate polymorphism by creating a function that can work with different shape objects to calculate and print their areas.

#Polymorphism allows different objects to be treated as instances of the same class or interface, even if they belong to different subclasses. In Python, this can be demonstrated by defining a function that can accept any object that follows a certain interface (like our Shape class with an area() method) and work with it, regardless of the specific subclass.
#Example of Polymorphism with Shape Objects:
from abc import ABC, abstractmethod
import math

# Abstract base class
class Shape(ABC):

    @abstractmethod
    def area(self):
        pass  # Abstract method, must be implemented by subclasses

# Circle subclass
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * (self.radius ** 2)

# Rectangle subclass
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# Triangle subclass
class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height

# Polymorphic function to calculate and print area
def print_area(shape):
    print(f"The area is: {shape.area()}")

# Example usage with different shape objects
circle = Circle(5)
rectangle = Rectangle(4, 6)
triangle = Triangle(3, 7)

# Using the polymorphic function to print the areas
print_area(circle)      # Will call Circle's area method
print_area(rectangle)   # Will call Rectangle's area method
print_area(triangle)    # Will call Triangle's area method

#Explanation:
#Polymorphic Function print_area():

#The function print_area() accepts any object that implements the area() method (i.e., any subclass of Shape).
#It doesn't need to know the exact type of the shape; it simply calls shape.area(), and Python determines which area() method to invoke based on the object passed.
#Demonstrating Polymorphism:

#The print_area() function works with instances of different subclasses (Circle, Rectangle, Triangle), but treats them uniformly as objects that have an area() method.
#Each shape's specific area() method is called based on its class, demonstrating polymorphism.

#Output:

#The area is: 78.53981633974483
#The area is: 24
#The area is: 10.5
#Key Concept:
#Polymorphism allows functions like print_area() to handle different types of shapes (Circle, Rectangle, Triangle) without needing to know their specific class, as long as they implement the area() method defined in the abstract base class Shape.


```

```python
#10. Implement encapsulation in a `BankAccount` class with private attributes for `balance` and`account_number`. Include methods for deposit, withdrawal, and balance inquiry
#Encapsulation is a key concept in object-oriented programming that restricts direct access to an object's internal state and allows controlled interaction through methods. In Python, this can be achieved using private attributes (denoted with double underscores __) and providing public methods to interact with those attributes.

#Here's an implementation of the BankAccount class with private attributes for balance and account_number, along with methods for depositing, withdrawing, and checking the balance:

#Example:

class BankAccount:
    def __init__(self, account_number, initial_balance=0):
        self.__account_number = account_number  # Private attribute
        self.__balance = initial_balance        # Private attribute

    # Method to deposit money
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited {amount}. New balance: {self.__balance}")
        else:
            print("Deposit amount must be positive.")

    # Method to withdraw money
    def withdraw(self, amount):
        if amount > 0:
            if self.__balance >= amount:
                self.__balance -= amount
                print(f"Withdrew {amount}. New balance: {self.__balance}")
            else:
                print("Insufficient funds.")
        else:
            print("Withdrawal amount must be positive.")

    # Method to check the balance
    def check_balance(self):
        print(f"Your current balance is: {self.__balance}")

    # Getter for account number (read-only)
    def get_account_number(self):
        return self.__account_number


# Example usage
account = BankAccount("123456789", 1000)  # Initial balance of 1000
account.check_balance()                    # Check balance

account.deposit(500)                       # Deposit 500
account.withdraw(200)                      # Withdraw 200
account.withdraw(2000)                     # Attempt to withdraw more than balance

print(f"Account Number: {account.get_account_number()}")  # Get account number

#Key Concepts:
#Private Attributes (__balance and __account_number):

#These are denoted with double underscores to make them private and inaccessible from outside the class. This encapsulates the internal state of the class.
#Public Methods:

#deposit(), withdraw(), and check_balance() provide controlled access to modify and retrieve the private attributes.
#get_account_number() allows read-only access to the private __account_number.
#Encapsulation:

#The balance and account number are encapsulated, and the only way to interact with them is through the public methods, enforcing control over how they are modified.
#Output:

#Your current balance is: 1000
#Deposited 500. New balance: 1500
#Withdrew 200. New balance: 1300
#Insufficient funds.
#Account Number: 123456789
#Benefits of Encapsulation:
#Data Protection: Prevents direct modification of sensitive data like balance and account number.
#Controlled Access: Interactions with the data are controlled through methods, allowing validation (e.g., ensuring that deposit/withdrawal amounts are positive).
```

```python
#12. Create a decorator that measures and prints the execution time of a function.
#In Python, decorators are a powerful and flexible way to modify the behavior of a function or method. You can create a decorator to measure and print the execution time of a function using the time module. Below is an example of how to implement such a decorator:

#Example of a Timing Decorator:

import time

def timing_decorator(func):
    """Decorator that measures the execution time of a function."""
    def wrapper(*args, **kwargs):
        start_time = time.time()  # Record the start time
        result = func(*args, **kwargs)  # Call the original function
        end_time = time.time()  # Record the end time
        execution_time = end_time - start_time  # Calculate execution time
        print(f"Execution time of '{func.__name__}': {execution_time:.6f} seconds")
        return result  # Return the result of the original function
    return wrapper

# Example function to demonstrate the decorator
@timing_decorator
def example_function(n):
    """A simple function that sums numbers from 1 to n."""
    total = sum(range(1, n + 1))
    return total

# Call the example function
result = example_function(1000000)
print(f"Result: {result}")
#Explanation:
#Importing the time Module:

#The time module provides various time-related functions. We use it to capture the start and end times of the function execution.
#Creating the Decorator:

#The timing_decorator function takes another function func as its argument.
#Inside, it defines a wrapper function that will be executed instead of the original function.
#Measuring Execution Time:

#The wrapper function records the start time using time.time().
#It then calls the original function (func(*args, **kwargs)), allowing for any arguments and keyword arguments to be passed.
#After the function execution, it records the end time and calculates the execution duration.
#Printing the Execution Time:

#The decorator prints the execution time formatted to six decimal places.
#Returning the Result:

#Finally, the wrapper returns the result of the original function, allowing the original functionality to remain intact.
#Example Function:
#The example_function sums integers from 1 to n. When decorated with @timing_decorator, its execution time will be printed each time it is called.
#Output:
#When you run the above code, you might see output similar to the following:


#Execution time of 'example_function': 0.123456 seconds
#Result: 500000500000
#Conclusion:
#This decorator is a simple and effective way to measure and display the execution time of any function, allowing you to easily monitor performance without modifying the function's core logic.

```

```python
#13. Explain the concept of the Diamond Problem in multiple inheritance. How does Python resolve it?

#The Diamond Problem is a common issue in multiple inheritance scenarios where a class inherits from two classes that both inherit from a common superclass. This creates a diamond-shaped inheritance structure, leading to ambiguity about which superclass method or property should be inherited.

'''Visual Representation of the Diamond Problem

        A
       / \
      B   C
       \ /
        D
In this example:

Class A is the base class.
Classes B and C both inherit from A.
Class D inherits from both B and C.
If class D tries to call a method that is defined in class A, it becomes unclear whether D should inherit the method from B or from C, resulting in ambiguity.'''

#How Python Resolves the Diamond Problem
#Python resolves the Diamond Problem using the Method Resolution Order (MRO). The MRO determines the order in which classes are searched when looking for a method or attribute. Python uses the C3 Linearization algorithm (C3 superclass linearization) to create a linear order of classes'''

#MRO Example in Python
#Let's illustrate how Python handles the Diamond Problem with an example:


class A:
    def greet(self):
        return "Hello from A"

class B(A):
    def greet(self):
        return "Hello from B"

class C(A):
    def greet(self):
        return "Hello from C"

class D(B, C):
    pass

# Create an instance of D
d = D()

# Call the greet method
print(d.greet())  # Output: Hello from B
#Explanation of the MRO
#MRO Calculation:

#For class D, the MRO is determined as follows:
#Start with D, then look at its parent classes B and C, and finally the base class A.
#The MRO for class D will be: D -> B -> C -> A.
#Method Lookup:

#When d.greet() is called, Python first checks D, then B, finds the greet() method there, and calls it, returning "Hello from B".
#If B had not defined the greet() method, Python would check C next, and finally A.
#Checking the MRO:

#You can view the MRO of a class using the __mro__ attribute or the mro() method:

print(D.__mro__)  # Output: (<class '__main__.D'>, <class '__main__.B'>, <class '__main__.C'>, <class '__main__.A'>, <class 'object'>)
#Benefits of MRO
#Consistency: MRO provides a consistent order for method resolution, ensuring that the first class found in the hierarchy is used.
#Avoids Ambiguity: By defining a clear search order, Python eliminates the ambiguity that can arise in multiple inheritance scenarios.
#Conclusion
#The Diamond Problem highlights the complexities of multiple inheritance, but Python's use of MRO and the C3 linearization algorithm provides a clear, deterministic approach to method resolution. This ensures that even in complicated inheritance hierarchies, the behavior of the classes remains predictable.

```

```python
#14. Write a class method that keeps track of the number of instances created from a class
#We can keep track of the number of instances created from a class by using a class variable. A class method can be used to modify or access this variable whenever a new instance is created. Below is an implementation of a class that demonstrates this concept:

#Example Implementation:

class InstanceCounter:
    # Class variable to keep track of the number of instances
    instance_count = 0

    def __init__(self):
        # Increment the instance count each time a new object is created
        InstanceCounter.instance_count += 1

    @classmethod
    def get_instance_count(cls):
        """Class method to return the number of instances created."""
        return cls.instance_count


# Example usage
obj1 = InstanceCounter()
obj2 = InstanceCounter()
obj3 = InstanceCounter()

# Retrieve and print the number of instances created
print(f"Number of instances created: {InstanceCounter.get_instance_count()}")  # Output: 3
#Explanation:
#Class Variable instance_count:

#A class variable instance_count is defined to keep track of how many instances of the class have been created. It is shared among all instances of the class.
#Constructor __init__:

#In the constructor, we increment the instance_count variable each time a new instance is created.
#Class Method get_instance_count():

#The class method get_instance_count() is defined with the @classmethod decorator. It takes cls as its first parameter, which represents the class itself.
#This method returns the current value of instance_count.
#Example Usage:

#Three instances of InstanceCounter are created. Each time an instance is created, the instance count is incremented.
#Finally, we call the class method to get the total number of instances created.
#Output:
#When you run the above code, you will see the following output:


#Number of instances created: 3
#Summary:
#This implementation shows how to use a class variable and a class method to keep track of the number of instances created from a class, allowing you to access this information without needing an instance of the class.

```

```python
#15. Implement a static method in a class that checks if a given year is a leap year
#A static method in a class is a method that does not require access to instance or class variables. It can be called on the class itself, and it is typically used for utility functions that don't modify the state of the class or its instances.

#To implement a static method that checks if a given year is a leap year, you can use the following rules:

#A year is a leap year if it is divisible by 4.
#However, if the year is divisible by 100, it is not a leap year, unless it is also divisible by 400.
#Example Implementation:

class YearChecker:

    @staticmethod
    def is_leap_year(year):
        """Static method to check if a given year is a leap year."""
        if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
            return True
        else:
            return False


# Example usage
year = 2024
if YearChecker.is_leap_year(year):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")

year = 1900
if YearChecker.is_leap_year(year):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")

year = 2000
if YearChecker.is_leap_year(year):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
#Explanation:
#Static Method is_leap_year():

#The method is defined using the @staticmethod decorator, indicating that it does not require access to instance or class variables.
#It takes a single argument year and checks whether it is a leap year based on the defined rules.
#Leap Year Logic:

#The logic checks if the year is divisible by 4 but not by 100 (unless it is also divisible by 400).
#It returns True if the year is a leap year and False otherwise.
#Example Usage:

#The static method is called directly on the class YearChecker, demonstrating that it can be used without creating an instance of the class.
#Several years are checked, and the results are printed.
#Output:
#When WE run the code, the output will be:

#2024 is a leap year.
#1900 is not a leap year.
#2000 is a leap year.
#Summary:
#This implementation demonstrates how to create a static method in a class to check if a given year is a leap year. It highlights the utility of static methods for functions that operate independently of the instance or class state.
```
